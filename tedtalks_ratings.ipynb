## 2. Which talks provoked the most online discussion

```python
ted_df.sort_values('comments').tail()
```

```python
ted_df['Views_per_comments'] =  ted_df.views / ted_df.comments
```

```python
ted_df.sort_values('Views_per_comments')
```

## 3. Visulize the distribution of comments

```python
ted_df.loc[ted_df.comments < 1000, 'comments'].plot(kind='hist', bins=20)
```

## 4. Plot the number of talks that took place each year

```python
ted_df.info()
```

```python
ted_df.film_date
```

```python
ted_df['film_date_time'] = pd.to_datetime(ted_df.film_date, unit='s')
```

```python
ted_df.dtypes
```

```python
ted_df.film_date_time.dt.year.head()
```

```python
ted_df.film_date_time.dt.year.value_counts().sort_index().plot(kind='line', title='Number of talks each year')
```

## 5. What were the "best" events in TED history to attend

```python
ted_df.groupby('event').views.agg(['count', 'mean','sum']).sort_values('mean').tail()
```
